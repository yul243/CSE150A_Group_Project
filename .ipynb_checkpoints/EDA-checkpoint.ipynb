```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
BASE_DIR = os.getcwd()
DATA_PATH = os.path.join(BASE_DIR, "data", "blkjckhands.csv")

df = pd.read_csv(DATA_PATH)

df
```

```python
## The data looks fine, there are no null values.
df.info()
```

```python
df['blkjck'].value_counts()
```

```python
df['winloss'].value_counts()
```

```python
# I found that the 'winloss' and 'blkjck' columns are both limited to a few categories. 
# To reduce memory usage, I changed them to the category type, which will make us run faster when counting and querying.
df['winloss'] = df['winloss'].astype('category')
df['blkjck'] = df['blkjck'].astype('category')
```

```python
df.describe()
```

```python
# I found that the probability of card 3-5 being 0 for both the player and the dealer is very high, 
# which indicates that in most cases, the player and the dealer will not take more than two cards.
zero_counts = (df[['card3', 'card4', 'card5', 'dealcard3', 'dealcard4', 'dealcard5']] == 0).mean()
print(zero_counts)
```

```python
df[['plwinamt', 'dlwinamt']].describe()
```

```python
df.groupby('winloss')[['plwinamt', 'dlwinamt']].describe()
```

```python
wrong_win = df[(df['winloss'] == "Win") & (df['plwinamt'] == 0)]

wrong_loss = df[(df['winloss'] == "Loss") & (df['dlwinamt'] == 0)]

wrong_nowin = df[(df['winloss'] == "nowin") & ((df['plwinamt'] > 0) | (df['dlwinamt'] > 0))]

print(f"Number of incorrect rows where the player wins but the player does not win money: {len(wrong_win)}")
print(f"Number of incorrect rows where the player loses but the player does not win money: {len(wrong_loss)}")
print(f"Number of incorrect rows where the tie does not have a value of 0: {len(wrong_nowin)}")

```

```python
# Calculate the sum of the player's hand.
df['calculated_sumofcards'] = df[['card1', 'card2', 'card3', 'card4', 'card5']].sum(axis=1, skipna=True)

# Calculate the total value of the dealer's hand.
df['calculated_sumofdeal'] = df[['dealcard1', 'dealcard2', 'dealcard3', 'dealcard4', 'dealcard5']].sum(axis=1, skipna=True)

player_sum_mismatch = df[df['calculated_sumofcards'] != df['sumofcards']]
dealer_sum_mismatch = df[df['calculated_sumofdeal'] != df['sumofdeal']]

print(f"The number of rows in which the player's `sumofcards` calculation is incorrect is: {len(player_sum_mismatch)}")
print(f"The number of rows in which the dealer's `sumofdeal` calculation is incorrect is: {len(dealer_sum_mismatch)}")
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df.groupby('ply2cardsum')['winloss'].apply(lambda x: (x == "Win").mean()).index,
             y=df.groupby('ply2cardsum')['winloss'].apply(lambda x: (x == "Win").mean()),
             marker='o', color='green')
plt.xlabel('Player Initial 2-Card Sum')
plt.ylabel('Win Rate')
plt.title('Player Win Rate by Initial 2-Card Sum')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df.groupby('sumofdeal')['winloss'].apply(lambda x: (x == "Win").mean()).index,
            y=df.groupby('sumofdeal')['winloss'].apply(lambda x: (x == "Win").mean()),
            color='blue')
plt.xlabel('Dealer Hand Total')
plt.ylabel('Player Win Rate')
plt.title('Player Win Rate Against Dealer Hand Total')
plt.show()
```

```python
# The above chart is logical when the dealer's hand points are 8-21.
# However, when the points are 22-26, the player's winning rate is not 100%.
# Therefore, I checked whether the player would go bust first when the dealer's hand points are greater than 21.
# The check found that this is indeed the case, and the reason the player loses is because they have gone bust before.
df[(df['sumofdeal'] > 21) & (df['winloss'] == "Loss")]['plybustbeat'].value_counts()
```
